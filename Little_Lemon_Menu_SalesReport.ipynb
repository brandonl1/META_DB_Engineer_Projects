```python
#Task 4 - Create a stored procedure named BasicSalesReport that returns the following statistics: 
#Total sales
#Average sale
#Minimum bill paid
#Maximum bill paid

import mysql.connector as connector

# Connect to the MySQL database
connection = connector.connect(user="root", password="Kobe!1", database="little_lemon_db")
cursor = connection.cursor()

# Drop the procedure if it exists.do this to check so you don't get duplicate create errors
try:
    cursor.execute("DROP PROCEDURE IF EXISTS BasicSalesReport")
    print("Previous 'BasicSalesReport' procedure dropped.")
except connector.Error as err:
    print("Error dropping procedure: {}".format(err))

# Create the stored procedure
basic_sales = """
CREATE PROCEDURE BasicSalesReport()
BEGIN
    SELECT
        SUM(BillAmount) AS TotalSales,
        AVG(BillAmount) AS AverageSale,
        MIN(BillAmount) AS MinBillPaid,
        MAX(BillAmount) AS MaxBillPaid
    FROM Orders;
END
"""
try:
    cursor.execute(basic_sales)
    print("Stored procedure 'BasicSalesReport' created successfully.")
except connector.Error as err:
    print("Error creating procedure: {}".format(err))

# Call the stored procedure
try:
    cursor.callproc('BasicSalesReport')

    # Fetch and print the results
    for result in cursor.stored_results():
        for row in result.fetchall():
            print(f"Total Sales: {row[0]}, Average Sale: {row[1]}, Minimum Bill Paid: {row[2]}, Maximum Bill Paid: {row[3]}")
except connector.Error as err:
    print("Error calling procedure: {}".format(err))

# Close the cursor and connection
cursor.close()
connection.close()
```
